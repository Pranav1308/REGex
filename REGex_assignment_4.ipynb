# Scatter Plot

```python
plt.figure(figsize=(10,10))
plt.scatter(data.High,data.Low)
plt.show()
```

```python
sns.pairplot(data=data,vars=['Price(in dollars)','Open','Change%'])
plt.show()
```

# Histogram

```python
plt.hist(data['Change%'])
```

# Barplot

```python
plt.figure(figsize=(50,4))
sns.barplot(data['Vol.'], data['Low'])
```

# Lineplot

```python
plt.figure(figsize=(10,6))
plt.style.use('seaborn-darkgrid')
sns.lineplot(x = data['Price(in dollars)'], y =  data['Open'],linewidth=1.5,label='Open')
sns.lineplot(x = data['Price(in dollars)'], y =  data['High'],linewidth=1.5,label='High')
```

# Heatmap

```python
plt.figure(figsize=(10,8))
sns.heatmap(data.corr(),annot=True)
```
